```python
#importing librares
import pandas as pd
from datetime import datetime,timedelta
```

```python
#Reading in dataset in dataframe
zulo_bank = pd.read_csv('zulo_bank.csv')
```

```python
#Displaying headers and rows
display(zulo_bank.head())
```

```python
#checking for data cleaninless
zulo_bank.info()
```

```python
#Filling up missing values
zulo_bank.fillna({'LoanAmount':0.0, 'LoanType':'unknown', #'StartDate':'unknown','EndDate':'unknown',
                  'InterestRate':0.0},inplace = True)
```

```python
zulo_bank.info()
```

```python
#Converting to 1NF
#Splitting the FullName into FirstName and LastName
zulo_bank[['FirstName','LastName']] = zulo_bank['FullName'].str.split(expand =True)
zulo_bank.head(5)
```

```python
#converting from denormalized to normalized data 1NF to 2NF
#Customer table
customer = zulo_bank[['FirstName','LastName','Email','Phone']].copy().drop_duplicates().reset_index(drop=True)
customer['CustomerID'] = range(1,len(customer)+1)
customer = customer[['CustomerID','FirstName','LastName','Email','Phone']]
```

```python
customer.head(5)
```

```python
#Account table
accounts = zulo_bank[['AccountType','Balance','OpeningDate']].copy().drop_duplicates().reset_index(drop = True)
accounts['AccountID'] = range(1,len(accounts)+1)
accounts = accounts[['AccountID','AccountType','Balance','OpeningDate']]
```

```python
accounts.head(5)
```

```python
#Transaction table
transactions = zulo_bank[['TransactionType','Amount','TransactionDate']].copy().drop_duplicates().reset_index(drop=True)
transactions['TransactionID'] = range(1,len(transactions)+1)
transactions = transactions[['TransactionID','TransactionType','Amount','TransactionDate']]
```

```python
transactions.head(5)
```

```python
#Loan table
loans = zulo_bank[['LoanAmount','LoanType','StartDate','EndDate','InterestRate']].copy().drop_duplicates().reset_index(drop=True)
loans['LoanID'] = range(1,len(loans)+1)
loans = loans[['LoanID','LoanAmount','LoanType','StartDate','EndDate','InterestRate']]
```

```python
loans.head(5)
```

```python
#Zulo bank table
#merge operations
zulo_bank = zulo_bank.merge(customer, on =['FirstName','LastName','Email','Phone'], how= 'left')\
                     .merge(accounts, on= ['AccountType','Balance','OpeningDate'], how= 'left')\
                     .merge(transactions, on = ['TransactionType','Amount','TransactionDate'], how= 'left')\
                     .merge(loans, on = ['LoanAmount','LoanType','StartDate','EndDate','InterestRate'], how= 'left')\
                      [['CustomerID','AccountID','TransactionID','LoanID']]
```

```python
zulo_bank.head(5)
```

```python

#Date Dimension table
#Defining start and end time
start_date = datetime(2020,1,1)
current_date = datetime(2090,12,31)

#calculating number of days btw start and current time
num_days = (current_date - start_date).days

#Generating list of dates from start to current
date_list = [start_date + timedelta(days=x) for x in range (num_days + 1)]

#Ensure date_id matches the length of the datelist
Date = {'DateID' : [x for x in range (1, len(date_list) + 1)],'Date': date_list}

#Create dataframe
date_dim = pd.DataFrame(Date)
date_dim['Year'] = date_dim['Date'].dt.year
date_dim['Month'] = date_dim['Date'].dt.month
date_dim['Day'] = date_dim['Date'].dt.day
date_dim['Date'] = pd.to_datetime(date_dim['Date']).dt.date
```

```python
date_dim.head(10)
```

```python
#converting from denormalized to normalized data 2NF to 3NF
#Accounts table
accounts['OpeningDate'] = pd.to_datetime(accounts['OpeningDate']).dt.date
accounts =  accounts.merge(date_dim, left_on= 'OpeningDate', right_on= 'Date', how= 'inner')\
                    . rename(columns={'DateID':'OpeningDate_ID'})\
                    .reset_index(drop=True)\
                    [['AccountID','AccountType','Balance','OpeningDate_ID']]
```

```python
accounts
```

```python
#converting from denormalized to normalized data 2NF to 3NF
#Transaction table
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate']).dt.date
transactions =  transactions.merge(date_dim, left_on= 'TransactionDate', right_on= 'Date', how= 'inner')\
                    . rename(columns={'DateID':'TransactionDate_ID'})\
                    .reset_index(drop=True)\
                    [['TransactionID','TransactionType','Amount','TransactionDate_ID']]
```

```python
transactions
```

```python
#converting from denormalized to normalized data 2NF to 3NF
#Loans table
loans['StartDate'] = pd.to_datetime(loans['StartDate']).dt.date
loans['EndDate'] = pd.to_datetime(loans['EndDate']).dt.date
loans =  loans.merge(date_dim, left_on= 'StartDate', right_on= 'Date', how= 'inner')\
              .rename(columns={'DateID':'StartDate_ID'}) \
              .merge(date_dim, left_on= 'EndDate', right_on= 'Date', how= 'inner',suffixes = ('','_end'))\
              .rename(columns={'DateID':'EndDate_ID'})\
              .reset_index(drop=True)\
               [['LoanID','LoanType','LoanAmount','StartDate_ID','EndDate_ID','InterestRate']]
```

```python
#connect to the newly created (or existing) database to create schema and tables
def get_db_connection():
    try:
        conn = psycopg2.connect(
             host = 'localhost',
            database = 'zulo_bank',
            user = 'postgres',
            password = 'u0987',
            port = '5432'
        )
        return conn
        conn = get_db_connection()
    except psycopg2.Error as e:
        print("Error: Unable to connect to the database.")
        print(e)
        return None
```

```python
#Creating a schema and tables
def create_tables():
    conn = get_db_connection()
    cursor = conn.cursor()
    create_table_query = '''
                            CREATE SCHEMA IF NOT EXISTS zulo_bankdb;

                            DROP TABLE IF EXISTS zulo_bankdb.transactions CASCADE;
                            DROP TABLE IF EXISTS zulo_bankdb.accounts CASCADE;
                            DROP TABLE IF EXISTS zulo_bankdb.customer CASCADE;
                            DROP TABLE IF EXISTS zulo_bankdb.date_dim CASCADE;
                            DROP TABLE IF EXISTS zulo_bankdb.loans CASCADE;
                            DROP TABLE IF EXISTS zulo_bankdb.zulo_fact_table CASCADE;

                            CREATE TABLE IF NOT EXISTS zulo_bankdb.date_dim(
                                DateID SERIAL PRIMARY KEY,
                                Date VARCHAR(1000),
                                Year INTEGER,
                                Month INTEGER,
                                Day INTEGER

                            );

                            CREATE TABLE IF NOT EXISTS zulo_bankdb.transactions(
                                TransactionID SERIAL PRIMARY KEY,
                                TransactionType VARCHAR(100),
                                Amount FLOAT,
                                TransactionDate_ID INTEGER,
                                FOREIGN KEY (TransactionDate_ID) REFERENCES zulo_bankdb.date_dim(DateID)
                                

                            );
                            CREATE TABLE IF NOT EXISTS zulo_bankdb.accounts(
                                AccountID SERIAL PRIMARY KEY,
                                AccountType VARCHAR(100),
                                Balance FLOAT,
                                OpeningDate_ID INTEGER,
                                FOREIGN KEY (OpeningDate_ID) REFERENCES zulo_bankdb.date_dim(DateID)
                                

                            );
                            CREATE TABLE IF NOT EXISTS zulo_bankdb.customer(
                                CustomerID SERIAL PRIMARY KEY,
                                FirstName VARCHAR(50),
                                LastName VARCHAR(50),
                                Email VARCHAR(50),
                                Phone  VARCHAR(50)

                            );
                            CREATE TABLE IF NOT EXISTS zulo_bankdb.loans(
                                LoanID SERIAL PRIMARY KEY,
                                LoanType VARCHAR(100),
                                LoanAmount FLOAT,
                                StartDate_ID INTEGER,
                                EndDate_ID INTEGER,
                                InterestRate FLOAT,
                                FOREIGN KEY (StartDate_ID) REFERENCES zulo_bankdb.date_dim(DateID),
                                FOREIGN KEY (EndDate_ID) REFERENCES zulo_bankdb.date_dim(DateID)

                            );
                            CREATE TABLE IF NOT EXISTS zulo_bankdb.zulo_fact_table(
                                CustomerID INTEGER,
                                AccountID INTEGER,
                                TransactionID INTEGER,
                                LoanID INTEGER,
                                FOREIGN KEY (CustomerID) REFERENCES zulo_bankdb.customer(CustomerID),
                                FOREIGN KEY (AccountID) REFERENCES zulo_bankdb.accounts(AccountID),
                                FOREIGN KEY (TransactionID) REFERENCES zulo_bankdb.transactions(TransactionID),
                                FOREIGN KEY (LoanID) REFERENCES zulo_bankdb.loans(LoanID)
                                
                                
                            );

                        '''
    cursor.execute(create_table_query)
    conn.commit()
    cursor.close()
    conn.close()
```

```python
#creating a function to load the data
import csv

def load_data_from_csv(csv_path):
    conn = get_db_connection()
    cursor = conn.cursor()
    with open(csv_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            cursor.execute(
                '''INSERT INTO zulo_bankdb.date_dim (DateID, Date, Year,Month,Day)
                   VALUES(%s,%s,%s,%s,%s);''',
                   row
            )
    conn.commit()
    cursor.close()
    conn.close()
#Providing csv path to the file
csv_file_path = r'datasets\database_model\date_dim.csv'
load_data_from_csv(csv_file_path)
print('date_dim loaded succesfully')
```

```python
def load_data_from_csv(csv_path):
    conn = get_db_connection()
    cursor = conn.cursor()
    with open(csv_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            try:
                cursor.execute(
                '''INSERT INTO zulo_bankdb.zulo_fact_table (CustomerID, AccountID, TransactionID, LoanID)
                   VALUES(%s,%s,%s,%s);''',
                   row
                )
            except psycopg2.IntegrityError: # to catch foreign key violation
                conn.rollback() # roll back process and skip
            
            else:
                conn.commit() # commits once exceptions are handled
    cursor.close()
    conn.close()
#Providing csv path to the file
csv_file_path = r'datasets\database_model\zulo_bank.csv'
load_data_from_csv(csv_file_path)
print('zulo_bank loaded succesfully')
```
